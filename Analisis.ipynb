```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df_indicators[['Pop_Pct', 'Birth_Pct', 'Life_Pct',
                  'Death_Pct', 'Fer_Pct', 'Med_Pct']].corr(numeric_only=True)
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlación: % de Población, Nacimientos, Vda, Muertes, Fertilidad, Edad Promedio")
plt.show()
```

```python
df_population = pd.read_sql("SELECT * FROM population;", connection)
df_birth = pd.read_sql("SELECT * FROM birth_rate;", connection)


country = "COL"

# Filtrar la fila del país
country_population = df_population[df_population["id"] == country].iloc[0]
country_birth_rate = df_birth[df_birth["id"] == country].iloc[0]

# Extraer las columnas de años y sus valores
years = [col for col in df_population.columns if col.startswith("YR")]

population = country_population[years].values
birth_rate = country_birth_rate[years].values

# Convertir los años a formato numérico
years = [int(a.replace("YR", "")) for a in years]

plt.figure(figsize=(10, 5))
plt.plot(years, population, marker="", color="royalblue", linewidth=2)
plt.title(f"Variación de la población en el tiempo ({country})")
plt.xlabel("Año")
plt.ylabel("Población")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(years, birth_rate, marker="", color="red", linewidth=2)
plt.title(f"Variación de la tasa de natalidad en el tiempo ({country})")
plt.xlabel("Año")
plt.ylabel("Tasa de natalidad")
plt.grid(True)
plt.tight_layout()
plt.show()
```
